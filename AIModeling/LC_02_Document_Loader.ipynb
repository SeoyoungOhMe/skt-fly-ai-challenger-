```python
# CSV 파일 읽기
from langchain.document_loaders import CSVLoader
loader = CSVLoader(
    file_path = '/content/titanic3.csv', csv_args={
        'delimiter':',',
        'quotechar':'"',
        'fieldnames':['pclass','survived','name','sex','age','sibsp','parch','ticket','fare','cabin','embarked','boat','body','home.dest']
    }
)

data = loader.load()
```
